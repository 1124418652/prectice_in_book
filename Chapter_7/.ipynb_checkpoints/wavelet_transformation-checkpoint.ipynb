```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['TakaoPGothic']     # 显示中文字体
```

```python
t1 = np.linspace(0, 2, 100, endpoint=False)      # 时间坐标分段
t2 = np.linspace(2, 4, 100, endpoint=False)
t3 = np.linspace(4, 6, 100, endpoint=False)
y1 = np.sin(2 * np.pi * t1 )                     # 频率随时间改变，三个频率分别在不同的时间出现
y2 = np.sin(2 * np.pi * 2 * t2)
y3 = np.sin(2 * np.pi * 4 * t3)
y = np.concatenate([y1, y2, y3])        # 将三个频率的曲线合并

t4 = np.linspace(0, 6, 300)
y4 = np.sin(2 * np.pi * t4) + np.sin(2 * np.pi * 2 * t4)\
    + np.sin(2 * np.pi * 4 * t4)                 # 在所有时刻，三种频率都存在

fig = plt.figure(figsize = (8, 8))
axes = plt.subplots(2, 1)[1]                     # 添加两个子图
plt.subplots_adjust(0, 0, 1, 1, 0.5, 0.5)
axes[0].plot(t4, y)
axes[0].set_title(u"curve in time domain(frequency change)")
axes[0].set_xlabel("t/s")
axes[0].set_ylabel("magnitude")

axes[1].plot(t4, y4)
axes[1].set_title(u"curve in time domain(frequency not change)")
axes[1].set_xlabel("t/s")
axes[1].set_ylabel("magnitude")
```

在上图中，时域曲线的频率随时间改变，频率分别为 1，2，4 Hz

```python
yy1 = np.fft.fft(y)     # 快速傅里叶变换
yy2 = np.fft.fft(y4)
mag1 = np.abs(yy1)      # 幅度计算
mag2 = np.abs(yy2)
freq = np.linspace(0, 50, 300)
print(freq[np.argmax(mag1)])
plt.plot(np.linspace(0, 50, 300), mag2)
```

```python
from matplotlib.font_manager import fontManager    # 获取字体

fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111)
plt.subplots_adjust(0,0,1,1,0,0)
plt.xticks([])
plt.yticks([])
x, y = 0.05, 0.05

fonts = [font.name for font in fontManager.ttflist \
         if os.path.exists(font.fname) and os.stat(font.fname).st_size > 1e6]

fonts = set(fonts)
dy = (1.0 - y) / ((len(fonts) // 4) + (len(fonts)%4 != 0))

for font in fonts:
    t = ax.text(x, y + dy / 2, u"中文字体", {'fontname': font, 'fontsize': 14},\
               transform = ax.transAxes)
    ax.text(x, y, font, {'fontsize': 12}, transform = ax.transAxes)
    x += 0.25
    if x >= 1.0:
        y += 0.05
        x = 0.05
```
